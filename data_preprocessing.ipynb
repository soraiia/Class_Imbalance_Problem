### 1. dataset_38_sick.csv

```python
df = pd.read_csv('class_imbalance/dataset_38_sick.csv')

num_cols = df.select_dtypes(include='float64').columns
cat_cols = df.select_dtypes(exclude='float64').columns

df.info()
df.isna().sum()
```

```python
df[num_cols] = df[num_cols].fillna(df[num_cols].mean().round(2))

df = df.drop(columns=['TBG'])

for col in cat_cols:
    if df[col].isna().any():
        value_counts = df[col].value_counts(dropna=True)
        if not value_counts.empty:
            most_frequent = value_counts.idxmax()  
            df[col].fillna(most_frequent, inplace=True)

df.isna().sum()
```

```python

df['sex'] = df['sex'].map({'F': 0, 'M': 1}) 
df['on_thyroxine'] = df['on_thyroxine'].map({'f': 0, 't': 1})
df['query_on_thyroxine'] = df['query_on_thyroxine'].map({'f': 0, 't': 1})
df['on_antithyroid_medication'] = df['on_antithyroid_medication'].map({'f': 0, 't': 1})
df['sick'] = df['sick'].map({'f': 0, 't': 1})
df['pregnant'] = df['pregnant'].map({'f': 0, 't': 1})
df['thyroid_surgery'] = df['thyroid_surgery'].map({'f': 0, 't': 1})
df['I131_treatment'] = df['I131_treatment'].map({'f': 0, 't': 1})
df['query_hypothyroid'] = df['query_hypothyroid'].map({'f': 0, 't': 1})
df['query_hyperthyroid'] = df['query_hyperthyroid'].map({'f': 0, 't': 1})
df['lithium'] = df['lithium'].map({'f': 0, 't': 1})
df['goitre'] = df['goitre'].map({'f': 0, 't': 1})
df['tumor'] = df['tumor'].map({'f': 0, 't': 1})
df['hypopituitary'] = df['hypopituitary'].map({'f': 0, 't': 1})
df['psych'] = df['psych'].map({'f': 0, 't': 1})
df['TSH_measured'] = df['TSH_measured'].map({'f': 0, 't': 1})
df['T3_measured'] = df['T3_measured'].map({'f': 0, 't': 1})
df['TT4_measured'] = df['TT4_measured'].map({'f': 0, 't': 1})
df['T4U_measured'] = df['T4U_measured'].map({'f': 0, 't': 1})
df['FTI_measured'] = df['FTI_measured'].map({'f': 0, 't': 1})
df['TBG_measured'] = df['TBG_measured'].map({'f': 0, 't': 1})

df = pd.get_dummies(df, columns=['referral_source'])

df['Class'] = df['Class'].map({'negative': 0, 'sick': 1})
target = df['Class']
df.drop(columns=['Class'], inplace=True)
df['Class'] = target
```

```python
df.info()
df.to_csv(os.path.join(output_folder, 'dataset_38_sick.csv'), index=False)
```
